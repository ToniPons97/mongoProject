```python
import yelp_api as ya
import folium
```

```python
df_starbucks = ya.get_coords("Starbucks", "Los Angeles")
```

```python
df_schools = ya.get_coords("Schools", "Los Angeles")
df_web_design = ya.get_coords("Web Design", "Los Angeles")
```

```python
def generateBaseMap(default_location=[34.0397, -118.2703], default_zoom_start=12):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
df_Airport = ya.get_coords("Airport", "Los Angeles")
```

```python
df_Airport
```

```python
base_map = generateBaseMap()
base_map.add_child(folium.ClickForMarker(popup='Potential Location'))
for i in df_starbucks["Coordinates"]:
    folium.Marker(i[::-1], icon=folium.Icon(icon="cloud") ,popup="Starbucks").add_to(base_map)
for i in df_schools["Coordinates"]:
    folium.Marker(i[::-1], icon=folium.Icon(icon="cloud", color="red") ,popup="Schools").add_to(base_map)
for i in df_web_design["Coordinates"]:
    folium.Marker(i[::-1], icon=folium.Icon(icon="cloud", color="green") ,popup="Web Design").add_to(base_map)
base_map
```
